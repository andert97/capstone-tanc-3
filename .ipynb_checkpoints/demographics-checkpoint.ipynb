# Demographics

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_excel('./data/Wallingford2018.xlsx')
#data.iloc[:,27:36]
```

### Annual Income

```python
annual_income = data['What is your estimated annual income']
annual_income = annual_income.iloc[1:]
ax = annual_income.value_counts().plot(kind='barh', title = "Estimated Annual Income")
ax.set_xlabel("Income Type")
ax.set_ylabel("Count")
```

### Zip Code

```python
zip_code = data['What is your zip code']
zip_code = zip_code.iloc[1:]
ax = zip_code.value_counts().plot(kind='bar', title = "Zip Code")
ax.set_xlabel("Zip Code")
ax.set_ylabel("Count")
```

### Race/Ethnicity

```python
data_white = data['Unnamed: 32'].iloc[1:]
white = data_white.count()
data_other = data['Unnamed: 31'].iloc[1:]
other = data_other.count()
data_latino = data['Unnamed: 30'].iloc[1:]
latino = data_latino.count()
data_hawaiian = data['Unnamed: 29'].iloc[1:]
hawaiian = data_hawaiian.count()
data_black = data['Unnamed: 28'].iloc[1:]
black = data_black.count()
data_asian = data['Unnamed: 27'].iloc[1:]
asian = data_asian.count()
df = pd.DataFrame({'lab':['White', 'Asian', 'Black', 'Hawaiian/Pacific Islander', 'Hispanic/Latino', 'Other'], 'val':[white, asian, black, hawaiian, latino, other]})
ax = df.plot.barh(x='lab', y='val', rot=0)
ax.set_xlabel("Ethnicity")
ax.set_ylabel("Count")
```
