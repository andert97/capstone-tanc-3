### Health & Wellness

#### Improved Physical Health

```python
import numpy as np
import matplotlib.pyplot as plt

take_better_care = data['TakeCareOfHealth']
take_better_care = take_better_care.replace(-1, 0)
take_better_care_counts = take_better_care.value_counts().rename_axis('unique_values').to_frame('count').sort_values(by=['unique_values'])
take_better_care_counts['x_values'] = take_better_care_counts.index


more_energy = data['HaveMoreEnergy']
more_energy = more_energy.replace(-1, 0)
more_energy_counts = more_energy.value_counts().rename_axis('unique_values').to_frame('count').sort_values(by=['unique_values'])
more_energy_counts['x_values'] = more_energy_counts.index

more_active = data['MoreActive']
more_active = more_active.replace(-1, 0)
more_active_counts = more_active.value_counts().rename_axis('unique_values').to_frame('count').sort_values(by=['unique_values'])
more_active_counts['x_values'] = more_active_counts.index

p1 = plt.bar(take_better_care_counts['x_values'], take_better_care_counts['count'], width = 0.5)
p2 = plt.bar(more_energy_counts['x_values'], more_energy_counts['count'], width = 0.5, bottom = take_better_care_counts['count'])
p3 = plt.bar(more_energy_counts['x_values'], more_energy_counts['count'], width = 0.5, bottom = more_energy_counts['count'])
plt.legend((p1[0], p2[0], p3[0]), ('Take Care of Health', 'More Energy', 'More Active'))
```

#### Improved Outlook

```python
feel_happier = data['FeelHappier']
feel_happier = feel_happier.replace(-1, 0)
feel_happier_counts = feel_happier.value_counts().rename_axis('unique_values').to_frame('count').sort_values(by=['unique_values'])
feel_happier_counts['x_values'] = feel_happier_counts.index

look_forward = data['SomethingToLookForwardTo']
look_forward = look_forward.replace(-1, 0)
look_forward_counts = look_forward.value_counts().rename_axis('unique_values').to_frame('count').sort_values(by=['unique_values'])
look_forward_counts['x_values'] =look_forward_counts.index

learned = data['LearnedAboutServices']
learned = learned.replace(-1, 0)
learned_counts = learned.value_counts().rename_axis('unique_values').to_frame('count').sort_values(by=['unique_values'])
learned_counts['x_values'] = learned_counts.index


p1 = plt.bar(feel_happier_counts['x_values'], feel_happier_counts['count'], width = 0.5)
p2 = plt.bar(look_forward_counts['x_values'], look_forward_counts['count'], width = 0.5, bottom = feel_happier_counts['count'])
p3 = plt.bar(learned_counts['x_values'], learned_counts['count'], width = 0.5, bottom = look_forward_counts['count'])
plt.legend((p1[0], p2[0], p3[0]), ('Feel Happier', 'Look Forward to Each Day', 'Learned About Services'))
```

### Social And Rec Activities

```python
volunteer_work = data['DoMoreVolunteerWork']
volunteer_work = volunteer_work.replace(-1, 0)
volunteer_work_counts = volunteer_work.value_counts().rename_axis('unique_values').to_frame('count').sort_values(by=['unique_values'])
volunteer_work_counts['x_values'] = volunteer_work_counts.index

see_friends = data['SeeFriendsMore']
see_friends = see_friends.replace(-1, 0)
see_friends_counts = see_friends.value_counts().rename_axis('unique_values').to_frame('count').sort_values(by=['unique_values'])
see_friends_counts['x_values'] = see_friends_counts.index


p1 = plt.bar(volunteer_work_counts['x_values'], volunteer_work_counts['count'], width = 0.5)
p2 = plt.bar(see_friends_counts['x_values'], see_friends_counts['count'], width = 0.5, bottom = volunteer_work_counts['count'])
plt.legend((p1[0], p2[0]), ('Feel Happier', 'Look Forward to Each Day'))
```
